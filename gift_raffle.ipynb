```python
import random
import ipywidgets as widgets
from IPython.display import display

def assign_gift(b, gift_assignments):
    giver = b.description
    if giver in gift_buyers:
        random.shuffle(gift_recipients)
        remaining_recipients = [recipient for recipient in gift_recipients if recipient != giver]
        if remaining_recipients:
            recipient = remaining_recipients[0]
            gift_assignments[giver] = recipient
            gift_recipients.remove(recipient)
            gift_buyers.remove(giver)
            display_gift_assignments(gift_assignments)
        else:
            with result_output:
                print("No available recipients left.")
    else:
        with result_output:
            print(f"{giver} has already made a selection")

def display_gift_assignments(gift_assignments):
    result_output.clear_output()
    with result_output:
        for giver, recipient in gift_assignments.items():
            print(f"{giver} buys a present for {recipient}")
```
